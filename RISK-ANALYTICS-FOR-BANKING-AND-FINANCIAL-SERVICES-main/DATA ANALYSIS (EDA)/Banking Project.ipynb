# Numerical Analysis

## In the Numerical Analysis will perform Univariate Analysis with Visualization 

```python
for i, col in enumerate (df[['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']]):
    plt.figure(figsize=[3,3])
    sns.histplot(df[col],kde=True)
    plt.title(col)
    plt.show()
```

# Heatmaps

```python
col_heat = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']
correlation = df[col_heat].corr()

plt.figure(figsize=[8,8])
sns.heatmap(correlation,annot=True,cmap='crest')
plt.title("correlation")
plt.show()
```

#### Derived Insight:
##### The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts" and "Saving Accounts" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.

# Plotting regplot for checking relationships:

```python
dff = [
    ('Bank Deposits', 'Checking Accounts'),
    ('Bank Deposits', 'Saving Accounts'),
    ('Checking Accounts', 'Saving Accounts'),
]


for x_col, y_col in dff:
    plt.figure(figsize=(4, 4))
    sns.regplot(
        data=df,
        x=x_col,
        y=y_col,
        scatter_kws={'alpha': 0.4}, 
        line_kws={'color': 'red'} 
    )
    plt.title(f'Relationship between {x_col} and {y_col}', fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.show()
```

#### Derived Insight:
##### The high correlation, we can see in this 3 graphs
